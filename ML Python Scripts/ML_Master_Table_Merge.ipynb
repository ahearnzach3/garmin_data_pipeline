# 2. Merge Dataframes

```python
from functools import reduce

# Define List of dataframes
dfs = [atl, maxmet, racepred, running, sleep, training_hist, uds]

# Perform the merge on 'calendarDate' for all dataframes
combined_df = reduce(lambda left, right: pd.merge(left, right, on='calendarDate', how='outer'), dfs)

# View Merged Dataframe
combined_df.head()
```

# 3. Clean Merged Dataframe

```python
# View shape
print(combined_df.shape) ## (281, 75)

# Count null values in each column
null_counts = combined_df.isna().sum()

# Display the null counts
## print(null_counts)
```

```python
# View rows where trainingStatus is null
## combined_df[['calendarDate','trainingStatus']]

# Drop rows where trainingStatus is null
combined_df_cleaned = combined_df[combined_df['trainingStatus'].notna()]

# View Dataframe
combined_df_cleaned
```

```python
# View shape
print(combined_df_cleaned.shape) ## (239, 75)

# Count null values in each column
null_counts = combined_df_cleaned.isna().sum()

# Display the null counts
## print(null_counts)
```

```python
# Fill missing 'vo2MaxValue' and 'maxMet' values with the value from the preceeding row
combined_df_cleaned.loc[:, 'vo2MaxValue'] = combined_df_cleaned['vo2MaxValue'].ffill()
combined_df_cleaned.loc[:, 'maxMet'] = combined_df_cleaned['maxMet'].ffill()

# Drop columns from 'running' dataframe that do not make sense to have with all other data
cols_to_drop = ['Activity_Type','Calories','Avg_HR','Max_HR'
                ,'Aerobic_TE','Avg_Pace','Best_Pace','Total_Ascent'
                ,'Total_Descent','Min_Elevation','Max_Elevation'
                ,'Distance_Group'
               ]

combined_df_cleaned = combined_df_cleaned.drop(cols_to_drop, axis=1)

# View Dataframe
combined_df_cleaned.head()
```

```python
# Rename 'Time' to 'Run_Time'
combined_df_cleaned.rename(columns={'Time': 'RunTime'}, inplace=True)
combined_df_cleaned.head()
```

```python
# View shape
print(combined_df_cleaned.shape) ## (239, 63)

# Count null values in each column
null_counts = combined_df_cleaned.isna().sum()

# Display the null counts
## print(null_counts)
```

```python
# Replace NaN values in 'Distance' with 0
combined_df_cleaned.loc[:, 'Distance'] = combined_df_cleaned['Distance'].fillna(0.00)

# Replace NaN values in 'RunTime' with 0
combined_df_cleaned['RunTime'] = combined_df_cleaned['RunTime'].fillna(pd.Timedelta(0))

# View Dataframe
combined_df_cleaned.head()
```

```python
# Assess Datatypes
## combined_df_cleaned.dtypes
```

```python
# List of columns to convert
time_columns = ['raceTime5K', 'raceTime10K', 'raceTimeHalf', 'raceTimeMarathon', 'RunTime']

# Convert each column to timedelta
for col in time_columns:
    combined_df_cleaned[col] = pd.to_timedelta(combined_df_cleaned[col])
```

```python
# Assess Datatypes
## combined_df_cleaned.dtypes
```

```python
# Create RunDay binary field
combined_df_cleaned['IsRunDay'] = combined_df_cleaned['Distance'].apply(lambda x: 1 if x > 0.0 else 0)

# Convert 'IsRunDay' to Object Datatype
combined_df_cleaned['IsRunDay'] = combined_df_cleaned['IsRunDay'].astype('object')
combined_df_cleaned.head()
```

```python
# Drop other columns that I think will be of little value for the ML Model
cols_to_drop = ['floorsAscendedInMeters','floorsDescendedInMeters','averageMonitoringEnvironmentAltitude','fitnessLevelTrend']
combined_df_cleaned = combined_df_cleaned.drop(cols_to_drop,axis=1)

# View Dataframe
combined_df_cleaned.head()
```

```python
# Convert 'IsVigorousDay' to object datatype
combined_df_cleaned['isVigorousDay'] = combined_df_cleaned['isVigorousDay'].astype('object')

## combined_df_cleaned.dtypes
```
